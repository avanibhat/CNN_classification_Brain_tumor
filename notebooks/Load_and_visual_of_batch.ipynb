Loading and Visualizing a batch of Images

```python
import matplotlib.pyplot as plt
import sys
import os
import torch
import torchvision
import numpy as np
sys.path.append(os.path.abspath("../src"))
from utils import get_dataloaders
```

1.Loading the dataset with the funciton get_dataloaders from the util.py file 
2. It loaded the MRI brain scan images from the Training and Testing folder.


```python
# Step 1: Load data
train_loader, test_loader, class_names = get_dataloaders("../data/brain_tumor_dataset/", batch_size=8)
print(f"Class labels: {class_names}")
```

```python
# Step 2: Get one batch
data_iter = iter(train_loader)
images, labels = next(data_iter)  # images: with each of shape [batch_size, 3, 224, 224]
```

```python

# Step 3: Helper to un-normalize and plot
def imshow(img_tensor):
    img_tensor = img_tensor * 0.5 + 0.5  # undo normalization ([-1,1] â†’ [0,1])
    npimg = img_tensor.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))  # Convert from [C, H, W] to [H, W, C]
    plt.axis('off')
    plt.show()
```

```python
# Step 4: Display first 8 images in a grid
imshow(torchvision.utils.make_grid(images))
```

```python
# Print class labels
print([class_names[label] for label in labels])
```
